```python
import pandas as pd
from happytransformer import HappyTextToText, TTTrainArgs
from datasets import Dataset


data = pd.read_csv("/content/personality.csv")


data["input"] = data["Persona"]
data["target"] = data["chat"]
data[["input", "target"]].to_csv("/content/personality_processed.csv", index=False)

dataset = Dataset.from_pandas(data[["input", "target"]])
data = data.drop(columns=["Persona", "chat", "Unnamed: 0"])

data.to_csv("/content/personality_processed.csv", index=False)





from happytransformer import HappyTextToText, TTTrainArgs

# Load GPT-Neo
happy_tt = HappyTextToText("EleutherAI/gpt-neo-125M")

args = TTTrainArgs(
    num_train_epochs=3,
    learning_rate=5e-5,
    batch_size=2,
    save_steps=100,
    eval_steps=100,
    output_dir="gptneo_finetune",
    fp16=False
)

happy_tt.train("/content/personality_processed.csv", args=args)





evaluation_data = {
    "input": [
        "Persona: I love reading books. What do you like to do?",
        "Persona: I enjoy hiking on weekends. What about you?"
    ],
    "target": [
        "Response: I enjoy watching movies in my free time.",
        "Response: I prefer spending my weekends exploring new places."
    ]
}

evaluation_df = pd.DataFrame(evaluation_data)

evaluation_df.to_csv("/content/evaluate.csv", index=False)

result = happy_tt.eval("eval.txt")
print("Evaluation Loss:", result.loss)

output = happy_tt.generate_text("What is the capital of Germany?")
print(output.text)
```

## We can use self-refinement as following:

```python
from happytransformer import HappyTextToText

personas = [
    "Persona: It's like there's something I should understand, but I can't grasp it.",
    "Persona: Sometimes I feel like I’m seeing everything from the wrong angle.",
    "Persona: I wonder if I missed something, but I’m not sure what.",
    "Persona: It's as if there’s a pattern, but it’s slipping through my fingers.",
    "Persona: I keep thinking I should do something, but I don't know what.",
    "Persona: Things seem so off, but I can't figure out why.",
    "Persona: I feel like there's an answer, but I don’t know the question.",
    "Persona: There's a feeling I can't put into words, but it’s there.",
    "Persona: I keep thinking about things, but I’m not sure they even matter.",
    "Persona: It’s like I’m looking at the pieces, but they don’t fit together."
]

for persona in personas:
    initial_response = happy_tt.generate_text(persona)

    refinement_prompt = f"Review your previous response and identify areas where you could offer more clarity, insight, or ask follow-up questions to deepen the conversation. Here is your initial response: {initial_response}"
    refined_response = happy_tt.generate_text(refinement_prompt)

    clarification_prompt = f"Based on the refined response, can you offer further insight or ask a follow-up question to help explore the topic more deeply?"
    clarification_response = happy_tt.generate_text(clarification_prompt)

    print(f"Persona: {persona}")
    print(f"Initial Response: {initial_response}")
    print(f"Refined Response: {refined_response}")
    print(f"Clarification: {clarification_response}")
    print("-" * 40)
```
